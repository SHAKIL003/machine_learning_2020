# Scrapping TMDB data using TMDB API

```python
import pandas as pd
import requests
```

```python
webpage = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=?')
```

```python
data = webpage.json()['results']
data
```

```python
df = pd.DataFrame(data)[['id', 'title', 'overview', 'release_date','popularity', 'vote_average', 'vote_count' ]]
```

```python
df.shape
```

```python
df_updated = pd.DataFrame()
```

```python
df_updated
```

```python
for i in range(1,461):
  webpage = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=?&language=en-US&page={}'.format(i))
  data = webpage.json()['results']
  df = pd.DataFrame(data)[['id', 'title', 'overview', 'release_date','popularity', 'vote_average', 'vote_count' ]]
  # df_updated = df_updated.append(df, ignore_index = True)
  df_updated = pd.concat([df_updated, df], ignore_index=True)
```
